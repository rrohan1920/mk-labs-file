```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

from google.colab import files
uploaded = files.upload()
Iris = list(uploaded.keys())[0]  # Gets the uploaded file name
df = pd.read_csv(Iris)
```

```python
print("the shape: ", df.shape)
print("the total number of colomns: ", len(df.columns))
```

```python
print("DataFrame Information:")
print(df.info())
```

```python
print("DataFrame description:")
print(df.describe())
```

```python
print("null values in the dataset")
print(df.isnull().sum())
```

```python
# total count of the unique values in the dataset
print("unique values in the dataset:", df.nunique())

# Values that are unique in column-vise
for col in df.columns:
    print(f"Unique values in column '{col}': {df[col].unique()}")
    print()
```

```python
df.hist(figsize=(12, 12), bins=20)
plt.show()
```

```python
sns.scatterplot(x=df['SepalLength'], y=df['SepalWidth'], hue=df['Species'])
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
sns.pairplot(df, hue="Species")
plt.show()
```

```python
# Select only numerical columns for correlation calculation
numerical_cols = df.select_dtypes(include=np.number).columns
print("Pairwise Correlation:")
print(df[numerical_cols].corr())

# Heatmap of correlation for numerical columns
plt.figure(figsize=(8, 6))
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df)
plt.title("Boxplot for Outlier Detection")
plt.show()
```
